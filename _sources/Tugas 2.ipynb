# **1. Equal Width Interval**

Equal Width Interval merupakan discretization yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar



```python
panjang_kelopak = dataset[["SepalLengthCm"]]
lebar_kelopak = dataset[["SepalWidthCm"]]
panjang_daunkelopak = dataset[["PetalLengthCm"]]
lebar_daunkelopak = dataset[["PetalWidthCm"]]
```

membuat fungsi **cut** yang digunakan untuk mencari interval menggunakan metode Equal-Width-Intervals

```python
def cut(col, k):
  intervals = pd.cut(dataset[col], k).value_counts().index.to_list()
  return [[interval.left, interval.right] for interval in intervals]
```

```python
def toCategory(list_interval, col):
  # get length interval
  length = len(list_interval)

  # sorting interval
  sort_interval = np.sort(list_interval, axis=0)

  # get category from interval
  categories = np.array([chr(65+i) for i in range(length)])[:, None]

  # Combine into interval data
  intervals = np.hstack((sort_interval, categories))

  # operate all data
  newCol = []
  for i, row in dataset.iterrows():
    d = row[col]
    for interval in intervals:
      if d >= interval[0].astype(float) and d <= interval[1].astype(float):
        newCol.append(interval[2])
        break

  # return new column category
  return np.array(newCol, dtype=str)
```

mencari interval dengan membaginya menjadi 3 bagian

```python
interval_panjang_kelopak = cut("SepalLengthCm", 3)
interval_lebar_kelopak = cut("SepalWidthCm", 3)
interval_panjang_daunkelopak = cut("PetalLengthCm", 3)
interval_lebar_daunkelopak = cut("PetalWidthCm", 3)

print("Interval_panjang_kelopak = ", interval_panjang_kelopak)
print("Interval_lebar_kelopak = ", interval_lebar_kelopak)
print("Interval_panjang_daunkelopak = ", interval_panjang_daunkelopak)
print("Interval_lebar_daunkelopak = ", interval_lebar_daunkelopak)
```

```python
panjang_kelopak["category"] = toCategory(interval_panjang_kelopak, "SepalLengthCm")
lebar_kelopak["category"] = toCategory(interval_lebar_kelopak, "SepalWidthCm")
panjang_daunkelopak["category"] = toCategory(interval_panjang_daunkelopak, "PetalLengthCm")
lebar_daunkelopak["category"] = toCategory(interval_lebar_daunkelopak, "PetalWidthCm")
```

# **2. Equal Frequency Intervals**

Equal Frequency Intervals merupakan discretization yang membagi data numerik menjadi beberapa kelompok dengan jumlah anggota yang kurang lebih sama besar.

Pandas menyediakan method qcut untuk mencari nilai interval dari Equal_Frequency Intervals




```python
def qcut(col, k):
  intervals = pd.qcut(dataset[col], k).value_counts().index.to_list()
  return [[interval.left, interval.right] for interval in intervals]
```

mencari interval dengan membaginya menjadi 3 bagian

```python
q_interval_panjang_kelopak = qcut("SepalLengthCm", 3)
q_interval_lebar_kelopak = qcut("SepalWidthCm", 3)
q_interval_panjang_daunkelopak = qcut("PetalLengthCm", 3)
q_interval_lebar_daunkelopak = qcut("PetalWidthCm", 3)

print("Interval_panjang_kelopak = ", q_interval_panjang_kelopak)
print("Interval_lebar_kelopak = ", q_interval_lebar_kelopak)
print("Interval_panjang_daunkelopak = ", q_interval_panjang_daunkelopak)
print("Interval_lebar_daunkelopak = ", q_interval_lebar_daunkelopak)
```

```python
#menampilkan hasil kategori

panjang_kelopak["category"] = toCategory(q_interval_panjang_kelopak, "SepalLengthCm")
lebar_kelopak["category"] = toCategory(q_interval_lebar_kelopak, "SepalWidthCm")
panjang_daunkelopak["category"] = toCategory(q_interval_panjang_daunkelopak, "PetalLengthCm")
lebar_daunkelopak["category"] = toCategory(q_interval_lebar_daunkelopak, "PetalWidthCm")
```

# **3. Entropy**

entropi adalah nilai yang menyatukan ukuran ketidakpastian (impurity) dan attribur sari suatu onjek data dalam satuan bit

pada peugasan ini akan menggunakan data panjang kelopak (sepal length) sembagi sampel perhitungan

```python
sampel = dataset [["SepalLengthCm"]]
sampel.describe ()
```

```python
#membuat kategori random untuk semua data

np.random.seed(0)
sampel["category"] = np.where(np.random.choice(2, sampel.shape[0]) < 1, "A", "B")
```

```python
print('Pengelompokan data uji dengan kategori random')
sampel
```

membuat fungsi **getOverCategory** yang digunakan untuk menghitung data keseluruhan yang nantinya digunakan untuk menghitung entropy

```python
def getOverCategory(col):
  group = sampel.loc[:, :].groupby("category").count()
  a = group.loc["A", col]
  b = group.loc["B", col]
  return (a, b, a+b)
```

fungsi splitter digunakan untuk membuat split antara value yang telah ditentukan lalu mengembalikan data yang telah dipisahkan

```python
def splitter(value:float, col:str)->tuple:
  # get data less and greater from value
  less = sampel[sampel[col] <= value]
  greater = sampel[sampel[col] > value]

  # calculate into category for each data
  less_group = less.loc[:, :].groupby("category").count()
  greater_group = greater.loc[:, :].groupby("category").count()

  # get value based on category
  less_category_A = less_group.loc["A", col] 
  less_category_B = less_group.loc["B", col] 
  greater_category_A = greater_group.loc["A", col] 
  greater_category_B = greater_group.loc["B", col] 

  return (
      [less_category_A, less_category_B, less_category_A + less_category_B],
      [greater_category_A, greater_category_B, greater_category_A + greater_category_B]
  )
```

Rumus Mencari Entropy:
$$ Entropy (D_1) = - Σ _1^n\ pi \log_2 \ pi  $$


```python
def entropy (d):
  r1 = (d[0] / d[2]) * np.log2(d[0] / d[2])
  r2 = (d[1] / d[2]) * np.log2(d[1] / d[2])
  return np.sum([r1, r2]) * -1
```

```python
#membuat fungsi info

def info (d):
  r1 = (d[0][2] / sampel.shape [0]) * entropy (d[0])
  r2 = (d[1][2] / sampel.shape [0]) * entropy (d[1])
  return r1 + r2
```

Membuat fungsi gain untuk menghitung selisih antara entropy awal dengan yang baru

Rumus Gain:

$$ Gain(Enew) = (Einitial) - (Enew) $$

```python
def gain(Einitial, Enew):
  return Einitial - Enew
```

```python
D = getOverCategory("SepalLengthCm")
entropy_d = entropy(D)
print("D = ",D)
print("Entropy D = ", entropy_d)
```

Melakukan beberapa tes split untuk mencari hasil dan informasi yang terbaik



```python
#Tes Pertama : Split 1:4.4

split1 = splitter (4.4, "SepalLengthCm")
info_split1 = info (split1)

print("Gain = ",  gain (entropy_d, info_split1))
```

```python
#Tes Pertama : Split 2:5.5

split2 = splitter (5.5, "SepalLengthCm")
info_split2 = info (split2)

print("Gain = ",  gain (entropy_d, info_split2))
```

```python
#Tes Pertama : Split 3:7.0

split3 = splitter (7.0, "SepalLengthCm")
info_split3 = info (split3)

print("Gain = ",  gain (entropy_d, info_split3))
```
