```python
#CONSTAN VARIABLE

DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

CONTINGENCY_TABLE_VALUE ={
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0)
}
```

```python
def get_series (dataset, idx, series):
  return dataset.loc[(idx - DECREMENT_BY_ONE), series]
```

```python
def get_dissimilarity_dataset(df, series_index = [], series = []):
    first_series = get_series(df, series_index[0], series)
    second_series = get_series(df, series_index[1], series)
    data = pd.concat([first_series,second_series],axis=1)
    return data.T
```

$ \frac{r + s}{q + r + s}\ $
